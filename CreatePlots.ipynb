```python
def read_data(filename):
    counts = [0] * 256
    with open(filename) as file:
            for line in file.readlines():
                tokens = line.strip().split(":")
                counts[int(tokens[0])] = int(tokens[1])
    
    return counts
```

```python
import matplotlib.pyplot as plt
from os.path import dirname, join

def plot_figure(filename, counts):
    fig = plt.figure()
    axes = fig.add_axes([0,0,1,1])
    
    axes.bar(list(range(256)), counts, width=1.0)
    axes.set_title(filename)
    
    plt.savefig(join("figures", dirname(filename) + "_plot.jpg"), bbox_inches='tight')
    plt.show()
```

```python
from os import listdir
from random import choice

filenames = [
    "image/tree.tiff_228",
    "text/jsPlumb-1.4.0-RC1.js_263",
    "audio/4558_2518.wav.wav_19",
    "executable/g++.exe_44",
]
    
for filename in filenames:
    counts = read_data(join("preprocessed_data", filename + ".txt"))
    plot_figure(filename, counts)
```
